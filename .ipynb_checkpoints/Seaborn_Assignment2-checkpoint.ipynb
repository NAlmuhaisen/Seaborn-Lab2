# Q1: Import seaborn library

```python
# write your code here ^_^
import seaborn as sns
```

```python
tips_dataset = sns.load_dataset('tips')
```

# Q2: Create a histogram, where the x-axis represents the tip column
- Change the style to 'whitegrid'.
- Split the data out by the day column's categories. Hint: use hue argument.
- Handle the overlapping columns by switching the element argument over to step.

```python
sns.set_style("whitegrid")
sns.histplot(tips_dataset, x="tip", hue="day", element="step");
```

## Q2.1 Write your top three findings 
- The diagrams shows that tips collected during Saturday and Sunday (USA weekend) are higher than the rest of the days 
- The customers number increases during the (USA weekend) specially on Saturday.
- most of the tips are between 2 and 4 S.R. or USD

# Q3: [Optional] Create a displot, where the x-axis represents the day column
- Split the data out by the sex column's categories. Hint: use hue argument.
- Handle the overlapping columns by switching the multiple argument over to 'dodge'.
- Shrink the bars down using shrink argument to have a nice gap between each of them.
- Change the bars color to 'Set1'.
- Reset the hue order to ['Female', 'Male'].

```python
sns.set_palette('Set1')
sns.displot(tips_dataset, x="day", hue="sex", multiple='dodge',shrink=0.6, hue_order=['Female', 'Male']);
```

## Q3.1 Write your top three findings 
- The diagram shows that during Saturday and Sunday there are more male customers, which may indicate that men are more likely to go to resturants during weekends.
- The diagram indicates that during Thursday and Friday there are more female customers. 
- In general, there are more male customers than female.

# Q4: Create a ecdf plot, where the x-axis represents the tip column
- Change the style to 'darkgrid'.
- Change the font scale to 1.25.
- Split the data out by the time column's categories. Hint: use hue argument.
- Change the lines color to 'summer'.
- Change the line width to 3.

```python
sns.set(font_scale=1.25)
sns.set_style("darkgrid")
sns.ecdfplot(tips_dataset, x="tip", hue="time",palette='summer',linewidth=3);
```

## Q4.1 Write your top three findings 
- There are a positive relation between the tip and the proportion
- Looking ate the dinner and launch curves, we can tell that there  are no big differance, however,
- The dinner curve is higher which means that customers are more likely to give more tips during dinner time.

# Q6: Create a violin plot, where the x-axis represents the time column and the y-axis represents the tip column
- Split the data out by the sex column's categories. Hint: use hue argument.
- Split the violin so that each violin contains both sex column's categories. Hint: use split argument.
- Change the shape of the inner part of the violin to 'quartiles'.
- Change the violin color to 'Pastel1'.
- Reset the hue order to ['Female', 'Male'].
- Change the line width to 2.
- Change the legend's location to The upper left.

```python
sns.set_palette('Pastel1')
sns.violinplot( data=tips_dataset, x="time", y="tip", hue="sex", linewidth=2, inner='quartiles', split=True, hue_order=['Female', 'Male'] );
plt.legend(loc ='upper left')
```

## Q6.1 Write your top three findings 
- The diagram shows that potential outliers in tip during dinner service
- Higher tip amounts during dinner time compared to the lunch time
- male are higher tippers than female

# Q8: Create a correlation heatmap 
- Adjust the figure's size to be (8,5).
- Change the color map to 'RdBu'.
- Add annotations to the heatmap.
- Set the minimum value to be -1.
- Set the maximum value to be 1.

```python
plt.figure(figsize=[8,5])
corrHeatMap = tips_dataset.corr()
sns.heatmap(corrHeatMap, cmap='RdBu', annot= True,vmin=-1, vmax=1) 
plt.show()
```

## Q8.1 Write your top three findings 
- There is a positive relation between tip and the total bill.
- There relation between the total bill is mediam 
- the lowest correlation is the one between tip and the size which mean that even if we have more members on the table it doesn't mean that they will pay more tip
